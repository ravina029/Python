```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
L=list(np.arange(10))
```

```python
M=list(2*np.arange(10))
M
```

# Scatter plot

```python
# Scatter plot
plt.scatter(L,M,color='r') #X and Y are here in list form 
plt.title('scatter plot',fontsize=10)
plt.xlabel("L",fontsize=10)
plt.ylabel("M",fontsize=10)
plt.show()
```

```python
# Scatter plot
colors=['r','b','g','y','r','b','g','y','g','y']
size=[12,24,53,95,46,79,89,99,100,22]
plt.scatter(L,M,c=colors,s=size,alpha=0.9999,marker="*",edgecolor='y',linewidth=0.2) #alpha is used to change the transparancy
plt.title('scatter plot',fontsize=10)
plt.xlabel("L",fontsize=10)
plt.ylabel("M",fontsize=10)
plt.show()
```

```python
colors=[23,34,45,56,67,89,10,32,54,12]
size=[12,24,53,95,46,79,89,99,100,22]
plt.scatter(L,M,c=colors,cmap='viridis',s=size,alpha=1) #alpha is used to change the transparancy
plt.colorbar()
plt.title('scatter plot',fontsize=10)
plt.xlabel("L",fontsize=10)
plt.ylabel("M",fontsize=10)
plt.show()
```

```python
colors=[23,34,45,56,67,89,10,32,54,12]
size=[12,24,53,95,46,79,89,99,100,22]
plt.scatter(L,M,c=colors,cmap='BrBG',s=size,alpha=1) #alpha is used to change the transparancy
plt.scatter(L,M,color='r')
t=plt.colorbar()
t.set_label('colorbar',fontsize=12)
plt.title('scatter plot',fontsize=10)
plt.xlabel("L",fontsize=10)
plt.ylabel("M",fontsize=10)
plt.show()
```

# bar plot

```python
X=['python','C',"C++","Java"]
y=[100,38,90,99]
c=['r','y','g','b']
plt.bar(X,y,width=0.5,color=c,align='edge',edgecolor='yellow',linewidth=5,alpha=1,label='barplot')
plt.legend()
plt.xlabel('X',fontsize=24)
plt.ylabel('y',fontsize=24)
```

```python
X=['python','C',"C++","Java"]
y=[100,38,90,99]
z=[20,28,79,54]
width=0.2
plt.bar(X,y,width,color='r',align='edge',label='barplot_y')

plt.bar(X,z,width=0.5,color='b',align='edge',edgecolor='yellow',label='barplot_z')
plt.legend()
plt.xlabel('X',fontsize=24)
plt.ylabel('y',fontsize=24)
```

```python
X=['python','C',"C++","Java"]
y=[100,38,90,99]
z=[20,28,79,54]
p=np.arange(len(X))
width=0.2
p1=[j+width for j in p]
plt.bar(X,y,width,color='r',align='edge',label='barplot_y')

plt.bar(p1,z,width,color='b',align='edge',edgecolor='yellow',label='barplot_z')
plt.legend()
plt.xticks(p+width/2,X,rotation=45)
plt.xlabel('X',fontsize=12)
plt.ylabel('y',fontsize=12)
```

```python
X=['python','C',"C++","Java"]
y=[100,38,90,99]
z=[20,28,79,54]
p=np.arange(len(X))
width=0.2
p1=[j+width for j in p]
plt.barh(X,y,color='r',label='barplot_y')

plt.barh(p1,z,color='b',label='barplot_z')
plt.legend()

plt.xlabel('y',fontsize=12)
plt.ylabel('X',fontsize=12)
```

```python
#to show the percentage of the part
y=['python','C',"C++","Java"]
X=[100,38,90,99]
ex=[0.1,0.2,0,0]
plt.pie(X,labels=y,autopct="%0.2f%%",explode=ex)
plt.show()
```

```python
#Shadow paramter
y=['python','C',"C++","Java"]
X=[100,38,90,99]
ex=[0.1,0.2,0,0]
plt.pie(X,labels=y,autopct="%0.2f%%",explode=ex,shadow=True)
plt.show()
```

```python
#Setting radius
y=['python','C',"C++","Java"]
X=[100,38,90,99]
ex=[0.1,0.2,0,0]
plt.pie(X,labels=y,autopct="%0.2f%%",explode=ex,shadow=True,radius=0.8)
plt.show()
```

```python
#Setting radius
y=['python','C',"C++","Java"]
X=[100,38,90,99]
ex=[0.1,0.2,0,0]
plt.pie(X,labels=y,autopct="%0.2f%%",explode=ex,shadow=True,radius=0.8,labeldistance=0.9)
plt.show()
```

```python
#Setting the starting angle
y=['python','C',"C++","Java"]
X=[100,38,90,99]
ex=[0.1,0.2,0,0]
plt.pie(X,labels=y,autopct="%0.2f%%",explode=ex,shadow=True,radius=0.8,startangle=90)
plt.show()
```

```python
#Rotating the labels
y=['python','C',"C++","Java"]
X=[100,38,90,99]
ex=[0.1,0.2,0,0]
plt.pie(X,labels=y,autopct="%0.2f%%",explode=ex,shadow=True,radius=0.8,startangle=90,textprops={'fontsize':8},rotatelabels=True)
plt.legend(loc=1)
plt.show()
```
