# Fecting perticular row or rows from the dataframe
* iloc: search by the index position
* loc: search by index labels


```python
student_info={'name':['harry','tom','rancho','kabir','rishi'],
    'iq':[100,90,80,80,89],
             'marks':[90,80,75,70,89],
             'package':[20,10,19,18,20]}
st=pd.DataFrame(student_info)
print(st)
```

```python
st.set_index('name',inplace=True)
st
```

```python
#iloc : fectching a perticular row by index number
#fetching a single row, it will a series data
st.iloc[0] 
st.iloc[1]
st.iloc[2]
```

```python
#fecthing multiple rows: it will a dataframe containg the specified rows
st.iloc[0:3]
st.iloc[3:]
```

```python
#skip alternate rows
st.iloc[::2] 
```

```python
#loc:fetching rows using lable name
st.loc['tom']
```

```python
st.loc['tom':'rishi']
```

```python
st.loc['tom':'rishi':2]
```

```python
#fancy  indexing 
st.loc[['tom','rishi']]
```
